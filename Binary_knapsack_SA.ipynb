```python
import random
import math

#sa parameters
initial_temperature=100
cooling_rate=0.95
max_iterations=100

def acceptance_solution(old_fitness,new_fitness,temperature):
    if old_fitness<new_fitness:
        return 1
    else:
        return math.exp((new_fitness-old_fitness)/temperature)
    
def fitness(current,objects,values,capacity):
    total_weight=0
    total_value=0
    for i in range(len(objects)):
        total_weight += current[i]*objects[i]
        total_value += current[i]*values[i]
    if capacity<total_weight:
        return 0
    else:
        return total_value
    
def find_neighbor(current):
    neighbor=list(current)
    index=random.randint(0,len(current)-1)
    neighbor[index]=1-neighbor[index]
    
    return neighbor
    
def simulated_anneling(solution,objects,values,capacity):
    temperature=initial_temperature
    while temperature>1e-6:
        for i in range(max_iterations):
            neighbor=find_neighbor(solution)

            old_fitness=fitness(solution,objects,values,capacity)
            new_fitness=fitness(neighbor,objects,values,capacity)

            if acceptance_solution(old_fitness,new_fitness,temperature)>random.random():
                solution=neighbor

        temperature *= cooling_rate
    print("final solution: ",solution)
    print("total value: ",fitness(solution,objects,values,capacity))
    
#driver code
capacity=int(input("enter capacity of knapsack: "))
n=int(input("enter no. of objects: "))
objects=[]
values=[]

for i in range(n):
    print("weight:")
    objects.append(int(input()))
    print("corresponding value:" )
    values.append(int(input()))
    
solution=[]
#generate initial solution using random function
for i in range(len(objects)):
    solution.append(random.choice([0,1]))
    
simulated_anneling(solution,objects,values,capacity)    
```
